## Modeling: Can I predict house prices?

```python
X = df.drop(columns = ["medv","log_medv","log_tax","log_crim"])
X = sm.add_constant(X)
y = df[["medv"]]

no_trans_lin_model = sm.OLS(y,X).fit()
no_trans_lin_model.summary()
```

Can't compare magnitude between covariates, need to standardize. Can use t-statistics (standardized distance from null of 0) or pvalue to describe statistical significance. Can standardize the data before building the model to compare magnitude of covariates
